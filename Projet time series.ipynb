On va d'abord essayer de répliquer les résultats importants de l'article, en utilisant leur code pour l'algorithme de base, mais en codant nos propres simulations. Cela permet de vérifier qu'il n'y a pas de problème, et aide à clarifier la méthode.

On crée déjà une fonction pour simplifier le reste du code

```python
def compute_p_values(x, cov, k=None, beta=None):
    x = x.reshape(-1)
    n = len(x)
    p_value_list = []
    if k:
        n_segments = k + 1 
        sg_results, cost = dp.dp(x, n_segments)
    else:
        sg_results, cost = pelt.pelt(x, n, beta)
    p_value_list = compute_p_values_from_seg(x, sg_results, cov, None, k, beta)
    return sg_results[1:-1], p_value_list

def compute_p_values_from_seg(x, sg_results, cov, indice_to_compute=None, k=None, beta=None):
    x = x.reshape(-1)
    n = len(x)
    p_value_list = []
    if not indice_to_compute:
        indice_to_compute = list(range(1, len(sg_results) - 1))
    if k:
        n_segments = k + 1 
        for i in indice_to_compute:
            x_prime, eta_vec, eta_T_x = parametrize_x.parametrize_x(x, n, sg_results[i - 1], sg_results[i],
                                                                sg_results[i + 1], cov)
            opt_funct_set, opt_funct = dp_parametrized_x.dp_parametrized_x(x_prime, sg_results, n_segments, n)
            p_value_list.append(inference.inference(opt_funct_set, opt_funct, eta_vec, eta_T_x, cov, n))
    else:
        for i in indice_to_compute:
            x_prime, eta_vec, eta_T_x = parametrize_x.parametrize_x(x, n, sg_results[i - 1], sg_results[i],
                                                                                sg_results[i + 1],
                                                                                cov)
            opt_funct_set, opt_funct = pelt_perturbed_x.pelt_perturbed_x(x_prime, sg_results, n, beta)
            p_value = inference.inference(opt_funct_set, opt_funct, eta_vec, eta_T_x, cov, n)
            p_value_list.append(p_value)
    
    return p_value_list
    
```

On va regarder combien de changepoints conduisent à rejeter H0 parmi les changepoints corrects détectés.

```python
def correctly_detected(detected_cp, all_true_cp):
    """we say a cp is correctly detected if it's at distance 2 from a true cp"""
    for i in range(-3, 4):
        if detected_cp + i in all_true_cp:
            return True, detected_cp + i
    return False, None

                
```

```python
n_iter = 20
n=50
p_value_list = []
for _ in tqdm(range(n_iter)):
    signal, bkps = rpt.pw_constant(n_samples=n, n_bkps=7, noise_std=0.5)
    cov = 0.5 * np.eye(n)
    k = 9
    n_segments = k + 1
    x = signal.reshape(-1)
    sg_results, cost = dp.dp(x, n_segments)
    
    indice_to_compute = [i for i in range(1, n_segments) if correctly_detected(sg_results[i], bkps)]
    p_value_list.extend(compute_p_values_from_seg(x, sg_results, cov, indice_to_compute, k))
    
```

# 2) Influence de la covariance sur le résultat

La p-valeur calculée dépend de la covariance de la série temporelle. Pour des données réelles, il faut donc estimer cette covariance, ce qui n'est pas évident en présence de changepoints.

Calculer directement la covariance sur la série temporelle entière va bien sûr largement surestimer la covariance (à cause des changepoints):

```python
signal, bkps = rpt.pw_constant(n_samples=100, n_bkps=7, noise_std=1)
```

```python
np.std(signal)**2
```

Une autre possibilité est de d'abord calculer les changepoints, puis de calculer la covariance de chaque segments. Si on trouve plus de changepoints que la réalité, on va avoir tendance à sous-estimer la variance (et donc à rejeter trop d'hypothèse nulle). Pour résoudre ce problème, l'article calcule la covariance sur chaque segment, et prend le maximum (voir Appendix A.5). Il montre qu'avec cette méthode, le False Discovery Rate reste presque aussi bon que si on connaissait la variance (du moins sur un exemple simple).

Mais on peut aussi imaginer le problème inverse: si on trouve moins de changepoints que la réalité, on va avoir tendance à surestimer la variance (et donc à ne rejeter pas assez d'hypothèses nulles, i.e à avoir un faible power).

On va regarder ici si le problème se pose. Pour simplifier l'étude, on va se restreindre à des séries temporelles gaussiennes non corrélées.

On regarde d'abord comment évolue notre estimation de la covariance avec la vraie variance, le vrai nombre de changepoints, et k (le nombre de changepoints calculés par l'algorithme).

```python
k = 6
n_segments = k + 1
signal, bkps = rpt.pw_constant(n_samples=100, n_bkps=7, noise_std=2)
x = signal.reshape(-1)
n_segments = k + 1
sg_results, cost = dp.dp(x, n_segments)
```

```python
n_iter = 10
k_list = list(range(1, 25))
real_std = 1
n_bkps = 8
std_mean_mat = np.zeros((len(k_list), n_iter))
std_min_mat = np.zeros((len(k_list), n_iter))
std_max_mat = np.zeros((len(k_list), n_iter))
std_med_mat = np.zeros((len(k_list), n_iter))
for i, k in enumerate(k_list):
    for j in range(n_iter):
        signal, bkps = rpt.pw_constant(n_samples=100, n_bkps=n_bkps, noise_std=real_std)
        x = signal.reshape(-1)
        n_segments = k + 1
        sg_results, cost = dp.dp(x, n_segments)
        std_mean_mat[i][j] = np.mean([np.std(x[sg_results[l-1]:sg_results[l]]) for l in range(1, len(sg_results))])
        std_max_mat[i][j] = np.max([np.std(x[sg_results[l-1]:sg_results[l]]) for l in range(1, len(sg_results))])
        std_min_mat[i][j] = np.min([np.std(x[sg_results[l-1]:sg_results[l]]) for l in range(1, len(sg_results))])
        std_med_mat[i][j] = np.median([np.std(x[sg_results[l-1]:sg_results[l]]) for l in range(1, len(sg_results))])
    
std_mean = np.mean(std_mean_mat, 1)
std_min = np.mean(std_min_mat, 1)
std_max = np.mean(std_max_mat, 1)
std_med = np.mean(std_med_mat, 1)
```

# 3) Comment faire quand la série temporelle a de l'autocorrelation et qu'on ne connait pas la variance

```python
def correlated_time_series(n, noise_std, n_chgts, chgts_strenght, lengthcale_correlation):
    """generate a correlated time series with changepoints"""
    cov = np.zeros((n, n))
    cov += noise_std * np.eye(n)
    for i in range(n):
        for j in range(i+1, n):
            cov[i][j] = np.exp(-(i - j)**2/lengthcale_correlation)
            cov[j][i] = np.exp(-(i - j)**2/lengthcale_correlation)
    mean = np.zeros(n)
    old_mean=0
    for i in range(0, n, n // n_chgts):
        new_mean = old_mean + np.random.choice(chgts_strenght)
        mean[i:i +  n // n_chgts] = new_mean
        old_mean = new_mean
    x = np.random.multivariate_normal(mean, cov)
    return x, [i for i in range(0, n, n // n_chgts)], cov
    
```

```python
def deriv(x):
    return np.array([(x[i+1] - x[i-1])/2 if (i < len(x) - 1) else (x[i] - x[i-1]) for i in range(1, len(x)) ])
```
